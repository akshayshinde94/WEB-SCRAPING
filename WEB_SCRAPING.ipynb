```python
# IMPORT LIBRARIES

import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
# website url

url = 'https://quotes.toscrape.com/'
```

```python
r = requests.get(url)
r
```

```python
text = r.content
```

```python
text
```

```python
# reading the data html
# 'htmi.parser' helps to arrange contents just like HTML page.

html = BeautifulSoup(text,'html.parser')
```

```python
html
```

```python
# checking content in container class

con = html.find(class_ = 'container')
con
```

```python
C = con.find_all('div', class_ = 'quote')
C
```

```python
# extracting item form text class

for items in C:
    s = items.find('span', class_='text').text
    print(s)
```

```python
# extracting author name from AUTHOR Class

for author in  C:
    a = author.find('small', class_='author').text
    print(a)
```

```python
# extracting TAGS

for items in C:
    t = items.find('meta')
    tags = t.attrs['content']
    print(tags)
```

```python
#Creating variable Contents to save the contents in a list.

C = con.find_all('div', class_ = 'quote')

contents = []

for items in C:
    s = items.find('span', class_='text').text
    a = items.find('small', class_='author').text
    t = items.find('meta')
    tags = t.attrs['content']
    contents.append([s,a,tags])
    
print(contents)
```

```python
# repeating for extracting contents from all pages

contents =[]

for i in range (1,10):
    url = 'https://quotes.toscrape.com/'
    r = requests.get(url)
    text = r.content
    html = BeautifulSoup(text,'html.parser')
    con = html.find(class_ = 'container')
    C = con.find_all('div', class_ = 'quote')
    for items in C:
        s = items.find('span', class_='text').text
        a = items.find('small', class_='author').text
        t = items.find('meta')
        tags = t.attrs['content']
        contents.append([s,a,tags])

print(contents)
```

```python
#converting to DataFrame

df = pd.DataFrame(contents, columns = ['Quote', 'Author', 'Tag'])
df
```
